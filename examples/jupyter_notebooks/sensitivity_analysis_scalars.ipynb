```python
def non_dim_sensitivities(ds_init, ds_output_adj):
    non_dim_grads = {}
    fc = ds_output_adj['fc'].data
    print(fc)
    non_dim_grads['c_slide_initb'] = np.sum(np.abs(ds_output_adj['xx_c_slide_initb'].data)) * 8.5 / fc[0]
    # non_dim_grads['delta_tda_constb'] = np.sum(np.abs(ds_output_adj['xx_delta_tda_constb'].data)) * 3.74 / fc
    non_dim_grads['c_dis_dab'] = np.abs(np.sum(ds_output_adj['xx_c_dis_dab'].data)) * 15659.0 / fc[0]
    non_dim_grads['q_geob'] = np.sum(np.abs(ds_output_adj['xx_q_geob'].data * ds_init['q_geo'].data)) / fc[0]
    non_dim_grads['Hb'] = np.sum(np.abs(ds_output_adj['xx_Hb'].data * ds_init['H'].data)) / fc[0]
    non_dim_grads['gamma_sb'] = np.sum(np.abs(ds_output_adj['xx_gamma_sb'].data)) * 0.070458 / fc[0]
    non_dim_grads['s_statb'] = np.abs(np.sum(ds_output_adj['xx_s_statb'].data)) * 5.0 / fc[0]
    non_dim_grads['beta1b'] = np.abs(np.sum(ds_output_adj['xx_beta1b'].data)) * 2.73 / fc[0]
    non_dim_grads['beta2b'] = np.abs(np.sum(ds_output_adj['xx_beta2b'].data)) * 7.28 / fc[0]
    non_dim_grads['Pmaxb'] = np.abs(np.sum(ds_output_adj['xx_Pmaxb'].data)) * 0.6 / fc[0]
    non_dim_grads['mub'] = np.abs(np.sum(ds_output_adj['xx_mub'].data)) * 9.7155 / fc[0]
    # non_dim_grads['RHO_Ab'] = np.abs(np.sum(ds_output_adj['xx_RHO_Ab'].data)) * 3300.0 / fc[0]
    non_dim_grads['time_lag_asthb'] = np.abs(np.sum(ds_output_adj['xx_time_lag_asthb'].data)) * 3000.0 / fc[0]
    non_dim_grads['enh_fact_da_dummy2d_scalar'] = np.abs(np.sum(ds_output_adj['xx_enh_fact_da_dummy2d_scalarb'].data)) * 3.0 / fc[0]
    non_dim_grads['enh_intg_da_dummy2d_scalar'] = np.abs(np.sum(ds_output_adj['xx_enh_intg_da_dummy2d_scalarb'].data)) * 1.0 / fc[0]
    non_dim_grads['flex_rig_lithb'] = np.sum(np.abs(ds_output_adj['xx_flex_rig_lithb'].data)) * 1.e25 / fc[0]
    non_dim_grads['p_weert'] = np.abs(np.sum(ds_output_adj['xx_p_weertb'].data)) * 3.0 / fc[0]
    non_dim_grads['q_weert'] = np.abs(np.sum(ds_output_adj['xx_q_weertb'].data)) * 2.0 / fc[0]
    non_dim_grads['zsb'] = np.sum(np.abs(ds_output_adj['xx_zsb'].data * ds_init['zs'].data)) / fc[0]
    non_dim_grads['zl0b'] = np.sum(np.abs(ds_output_adj['xx_zl0b'].data * ds_init['zl0'].data)) / fc[0]
    non_dim_grads['zlb'] = np.sum(np.abs(ds_output_adj['xx_zlb'].data * ds_init['zl'].data)) / fc[0]
    non_dim_grads['zbb'] = np.sum(np.abs(ds_output_adj['xx_zbb'].data * ds_init['zb'].data)) / fc[0]
    non_dim_grads['temp_cb'] = np.sum(np.abs(ds_output_adj['xx_temp_cb'].data * ds_init['temp_c'].data)) / fc[0]
    non_dim_grads['vx_cb'] = np.sum(np.abs(ds_output_adj['xx_vx_cb'].data * ds_init['vx_c'].data)) / fc[0]
    non_dim_grads['vy_cb'] = np.sum(np.abs(ds_output_adj['xx_vy_cb'].data * ds_init['vy_c'].data)) / fc[0]
    non_dim_grads['vz_cb'] = np.sum(np.abs(ds_output_adj['xx_vz_cb'].data * ds_init['vz_c'].data)) / fc[0]
    non_dim_grads['age_cb'] = np.sum(np.abs(ds_output_adj['xx_age_cb'].data * ds_init['age_c'].data)) / fc[0] # * 3.156e+7 # year2sec
    non_dim_grads['delta_tdab'] = np.sum(np.abs(ds_output_adj['xx_delta_tdab'].data)) * 3.74 / fc[0]
    
    # Sum of all calculated non-dimensional sensitivities
    total_sum = sum(non_dim_grads.values())
    
    # Normalize each field to percentage and round to two decimal places
    non_dim_grads_percentage = {key: np.round((value / total_sum) * 100, 2) for key, value in non_dim_grads.items()}
    
    return non_dim_grads_percentage
```
